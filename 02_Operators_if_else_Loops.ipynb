# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

Arithmetic Operators



```python
print(5 + 6)
print(5 - 6)
print(5 * 6)
print(5 / 6)
print(5 // 6)
print(5 % 6)
print(5 ** 2)
```

Relational Operators

```python
print(4 > 5)
print(4 < 5)
print(4 >= 4)
print(4 <= 4)
print(4 == 4)
print(4 != 4)
```

Logical Operators

```python
# and

print(1 and 0)
print(1 and 1)
print(3 and 5) # print second
```

```python
# or

print(0 or 0)
print(0 or 1)
print(0 or 5)
print(3 or 5) # print first
```

```python
# not

print(not 0)
print(not 1)
print(not 5)
```

Bitwise Operators

```python
print(2 & 3)   # Bitwise AND
print(2 | 3)   # Bitwise OR
print(2 ^ 3)   # Bitwise XOR
print(~3)      # Bitwise NOT
print(4 >> 2)  # Right shift
print(5 << 2)  # Left shift
```

Assignment Operators

```python
a = 2

a += 2
print(a)

a -= 2
print(a)

a *= 2
print(a)

a /= 2
print(a)

a %= 2
print(a)
```

Membership Operators

```python
# in/not in

print(1 in [1,2,3,4,5,6])

print('D' not in 'Dhaka')
```

# If-else in Python

```python
# login program and indentation
# email -> nitish.campusx@gmail.com
# password -> 1234

userEmail = 'sony@gmail.com'
userPassword = '12345'

email = input('Enter Email: ')
password = input('Enter Password: ')

if userEmail == email and userPassword == password:
  print('Welcome')
elif userEmail == email and userPassword != password:
  print('Incorrect password')
  password = input('Enter password again: ')
  if userPassword == password:
    print('Welcome,finally!')
  else:
    print('Still incorrect password')
else:
  print('Password and email not match')
```

```python
# min of 3 number

a = int(input('First Number: '))
b = int(input('Second Number: '))
c = int(input('Third Number: '))

if a<b and a<c:
  print('Smallest is',a)
elif b<c:
  print('Smallest is',b)
else:
  print('smallest is',c)
```

```python
# menu driven calculator

fnum = int(input('Enter first number: '))
snum = int(input('Enter second number: '))

op = input('Enter the operation: ')

if op == '+':
  print(fnum+snum)
elif op == '-':
  print(fnum-snum)
elif op == '*':
  print(fnum*snum)
else:
  print(fnum/snum)
```

```python
# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
number = int(input('Enter the number: '))
print()
i=1
while i<11:
  print(number,'*',i,'=',number*i)
  i+=1
```

```python
# while loop with else

x = 1

while x < 3:
  print(x)
  x += 1

else:
  print('limit crossed')
```

```python
# Guessing game

# generate a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input('Guess a number between 1 to 100: '))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Wrong!guess higher')
  else:
    print('Wrong!guess lower')

  guess = int(input('Guess a number: '))
  counter += 1

else:
  print('Correct guess')
  print('Attempts',counter)


```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
num = int(input('Enter a number: '))

result = 0
fact = 1

for i in range(1,num+1):
  fact = fact * i
  result = result + (i/fact)

print(result)
```

### Pattern 1


*<br>
** <br>
***<br>


```python
rows = int(input('Enter number of rows: '))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print('*',end='')
  print()
```

### Pattern 2

1<br>
121<br>
12321<br>
1234321<br>

```python
rows = int(input('Enter number of rows: '))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')

  print()
```

# Loop Control Statement

- Break
- Continue
- Pass

```python
# Break demo

for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
# Continue demo

for i in range(1,10):
  if i == 5:
    continue
  print(i)
```
