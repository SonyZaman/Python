## 3. Implementing a More Complex `Atm` Class

```python
class Atm:

  # constructor(special function)->superpower ->
  def __init__(self):

    self.pin = ''
    self.balance = 0
    self.menu()

  def menu(self):
    user_input = input("""
    Hi how can I help you?
    1. Press 1 to create account
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    Choose your option: """)

    print()
    if user_input == '1':
      if self.balance == 0 and self.pin == '':
        self.create_account()
      else:
        print('Account already created')
        self.menu()
    elif user_input == '2':
      if self.balance == 0 and self.pin == '':
        print('Please create account first')
        self.menu()
      else:
        self.change_pin()
    elif user_input == '3':
      if self.balance == 0 and self.pin == '':
        print('Please create account first')
        self.menu()
      else:
        self.check_balance()
    elif user_input == '4':
      if self.balance == 0 and self.pin == '':
        print('Please create account first')
        self.menu()
      else:
        self.withdraw()
    else:
      exit()

  def create_account(self):
    user_pin = input('Enter your pin: ')
    self.pin = user_pin

    user_balance = int(input('Enter amount: '))
    self.balance = user_balance

    print('Pin created successfully')
    print('Amount deposited successfully')
    self.menu()

  def change_pin(self):
    old_pin = input('Enter old pin: ')

    if old_pin == self.pin:
      # let him change the pin
      new_pin = input('Enter new pin: ')
      self.pin = new_pin
      print('Pin changed successfully')
    else:
      print('Old password does not match')
    self.menu()

  def check_balance(self):
    user_pin = input('Enter your pin: ')
    if user_pin == self.pin:
      print('Your balance is ',self.balance)
    else:
      print('Password does not match')
    self.menu()

  def withdraw(self):
    user_pin = input('Enter the pin: ')
    if user_pin == self.pin:
      # allow to withdraw
      amount = int(input('Enter the amount: '))
      if amount <= self.balance:
        self.balance = self.balance - amount
        print('Withdrawl successful.balance is',self.balance)
      else:
        print('Not sufficient amount')
    else:
      print('Password does not match')
    self.menu()
```

## 5. Understanding the Magic Method

### `__init__` Method

The `__init__` method is a special method in Python classes. It is often referred to as the **constructor**. Its primary purpose is to initialize the attributes of an object when it is created.

**Key Characteristics:**
*   It is automatically called whenever a new object (instance) of the class is created.
*   The first parameter it receives is always `self`, which refers to the instance of the class being created.
*   You can define additional parameters to `__init__` to pass initial values for the object's attributes during its creation.

**Example:**
In the `Atm` class, `__init__` was used to set initial `pin` and `balance` values. In the `Temp` class below, it simply prints 'hello' to demonstrate its automatic execution upon object instantiation.

```python
class Temp:

  def __init__(self):
    print('hello')

obj = Temp()
```

### `__str__` Method

The `__str__` method is another special method in Python classes that defines how an object should be represented as a string. It is primarily used for creating a human-readable representation of an object, especially when you use functions like `print()` or `str()`.

**Key Characteristics:**
*   It should return a string.
*   It's called automatically when you use `print()` on an object or `str(object)`.
*   It's distinct from `__repr__`, which provides an unambiguous representation of an object, often used for debugging.

**Example:**
Below, we define a `Car` class with `__init__` to set its `make`, `model`, and `year`. The `__str__` method then defines how a `Car` object will be displayed when printed.

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def __str__(self):
        return f"Car: {self.year} {self.make} {self.model}"

# Create a Car object
my_car = Car("Toyota", "Camry", 2020)

# Print the Car object, which invokes __str__
print(my_car)

# Using str() directly
car_string = str(my_car)
print(car_string)
```

## 6. Operator Overloading with a `Fraction` Class

### Defining a `Fraction` class with custom arithmetic operations

```python
class Fraction:

  # parameterized constructor
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  # self -> current object (the object that calls the method)
  # other -> the second object passed as an argument (the object we operate with)
  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __sub__(self,other):
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __mul__(self,other):
    new_num = self.num*other.num
    new_den = self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    return '{}/{}'.format(new_num,new_den)

  def convert_to_decimal(self):
    return self.num/self.den




```

### Creating and using `Fraction` objects

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

### Performing arithmetic operations on `Fraction` objects

```python
# fr1 -> self -> current object (the object that calls the method)
fr1.convert_to_decimal()
```

```python
# fr1 -> self -> current object (the object that calls the method)
# fr2 -> other -> the second object passed as an argument (the object we operate with)
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
