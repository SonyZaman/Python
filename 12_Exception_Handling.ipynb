A simple text file named `sample.txt` and writes 'hello world' into it.

```python
# let's create a file
with open('sample.txt','w') as f:
  f.write('hello world')
```

A basic `try-except` block to handle a `FileNotFoundError` when trying to open `sample1.txt` (which doesn't exist).

```python
# try catch demo
try:
  with open('sample1.txt','r') as f:
    print(f.read())
except:
  print('sorry file not found')
```

How to catch multiple specific exceptions (`NameError`, `ZeroDivisionError`, `FileNotFoundError`) and a general `Exception` in a single `try-except` block.

```python
# catching specific exception
try:
  m=5
  print(m/0)

  f = open('sample1.txt','r')
  print(f.read())

  L = [1,2,3]
  L[100]
except NameError:
  print('variable not defined')
except ZeroDivisionError:
  print("can't divide by 0")
except FileNotFoundError:
  print('file not found')
except Exception as e:
  print(e)
```

The `else` block runs when the try block executes without any exception.

```python
# else
try:
  f = open('sample1.txt','r')
except FileNotFoundError:
  print("file couldn't find")
except Exception:
  print('something is wrong')
else:
  # if try have no error then else execute
  print(f.read())

```

The `finally` block, which always executes regardless of whether an exception occurred or not.

```python
# finally

try:
  f = open('sample1.txt','r')
except FileNotFoundError:
  print("file couldn't find")
except Exception:
  print('something is wrong')
else:
  print(f.read())
finally:
  print('Always print')
```

A `Bank` class with a `withdraw` method that uses `raise` to signal custom error conditions (e.g., negative amount, insufficient balance). The errors are then caught by an `except` block.

```python
# raise (throw) sends the error to the except (catch) block

class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise Exception('Amount cannot be -ve')
    if self.balance < amount:
      raise Exception("You don't have enough money")
    self.balance = self.balance - amount

obj = Bank(10000)
try:
  obj.withdraw(15000)
except Exception as e:
  print(e)
else:
  print(obj.balance)
```

Creating a custom exception class `MyException` and using it within the `Bank` class's `withdraw` method.

```python
# custom error class
class MyException(Exception):
  def __init__(self,message):
    print(message)

class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise MyException('Amount cannot be -ve')
    if self.balance < amount:
      raise MyException("You don't have enough money")
    self.balance = self.balance - amount

obj = Bank(10000)
try:
  obj.withdraw(5000)
except MyException as e:
  pass
else:
  print(obj.balance)
```

We create custom exceptions to get more control over error handling and to make the code more readable and maintainable.

```python
class SecurityError(Exception):

  def __init__(self,message):
    print(message)

  def logout(self):
    print('logout')

class Google:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise SecurityError('Trying logging from different device')
    if email == self.email and password == self.password:
      print('welcome')
    else:
      print('login error')



obj = Google('sony','sony@gmail.com','1234','android')

try:
  obj.login('sony@gmail.com','1234','windows')
except SecurityError as e:
  e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')

```
