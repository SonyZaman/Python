## Generator
Python generators are a simple way of creating iterators

***Basic Generator Example***

This function `gen_demo()` is a generator because it uses the `yield` keyword. When called, it returns a generator object, which is an iterator. Each call to `next()` on this object will resume execution from where it last yielded, producing the next value until `StopIteration` is raised.

```python
def gen_demo():

  yield "first statement"
  yield "second statement"
  yield "third statement"
  yield "fourth statement"

gen = gen_demo() # just object store, but not calling object

# function call from here and remember what yield prints last time
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

***Generator Exhaustion (`StopIteration`)***

Once a generator has yielded all its values, subsequent calls to `next()` will raise a `StopIteration` exception. This signals that there are no more items to produce.

```python
print(next(gen))
```

```python
def lazy_generator():
  print("Starting generator...")
  yield 1
  print("Generated 1, pausing...")
  yield 2
  print("Generated 2, pausing...")
  yield 3
  print("Generated 3, finishing...")

lazy_gen = lazy_generator()

print("Before first next() call.")
print(next(lazy_gen))
print("After first next() call.")

print("Before second next() call.")
print(next(lazy_gen))
print("After second next() call.")
```

***Iterating with a `for` Loop***

Generators are commonly iterated using `for` loops. The `for` loop automatically handles the `StopIteration` exception, making the code cleaner and more robust than manual `next()` calls.

```python
def gen_demo():

  yield "first statement"
  yield "second statement"
  yield "third statement"
  yield "fourth statement"

gen = gen_demo() # just object store, but not calling object

# function call from here, and remember what yield prints last time
for i in gen:
  print(i)
```

 `yield` vs. `return`

*   **`return`**: Terminates the function and sends a single value back to the caller. The function's local state is destroyed.
*   **`yield`**: Pauses the function's execution, sends a value back, and saves the function's local state. When the generator is resumed (e.g., via `next()`), it continues from where it left off and remember last value.

## Example

```python
def square(num):
  for i in range(1,num+1):
      yield i**2
```

***Retrieving Values with `next()`***

Here, we create a generator object `gen` from `square(10)` and manually retrieve each value using `next()`. Each call to `next()` executes the generator function until the next `yield` statement.

```python
gen = square(10)

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

***Handling `StopIteration`***

Attempting to call `next()` on an exhausted generator (one that has no more values to yield) will result in a `StopIteration` error, as shown here after all 10 squares have been retrieved.

```python
print(next(gen))
```

***Iterating with `for` Loop (Generator Object)***

This demonstrates creating a generator object `gen` and then iterating through it using a `for` loop. The `for` loop implicitly calls `next()` and handles the `StopIteration`.

```python
gen = square(10)

for i in gen:
  print(i)
```

***Iterating Directly (Generator Function Call)***

You can also directly iterate over the result of a generator function call within a `for` loop. In this case, a new generator object is created and consumed directly by the loop.

```python
for i in square(10):
  print(i)
```

## Implementing a Custom Iterable Class
This `my_range` class aims to replicate the behavior of Python's built-in `range()` function without using generators. It defines `__init__` to set the start and end values and `__iter__` to return an iterator object.

```python
class my_range:

  def __init__(self,start,end):
    self.start = start
    self.end = end

  def __iter__(self):
    return my_range_iterator(self)
```

***Implementing a Custom Iterator Class***

This `my_range_iterator` class is responsible for the actual iteration logic. It defines `__init__` to hold a reference to the iterable, `__iter__` to return itself (as it's its own iterator), and `__next__` to produce the next value or raise `StopIteration` when exhausted.

```python

class my_range_iterator:

  def __init__(self,iterable_obj):
    self.iterable = iterable_obj

  def __iter__(self):
    return self

  def __next__(self):

    if self.iterable.start>=self.iterable.end:
      raise StopIteration

    current = self.iterable.start
    self.iterable.start+=1
    return current
```

***Demonstrating `next()` with Custom Iterable***

Here, we create an instance of our custom `my_range` class and retrieve values one by one using `next()`, similar to how generators work. This illustrates the manual control over iteration.

```python
ran = my_range(1,11)

print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
print(next(ran))
```

***Custom Iterator Exhaustion (`StopIteration`)***

Just like generators, our custom iterator raises `StopIteration` when all elements have been yielded, indicating the end of the sequence.

```python
print(next(ran))
```

***Iterating with `for` Loop (Custom Iterable Object)***

This demonstrates using a `for` loop with an instance of our custom `my_range` class. The `for` loop internally calls `__iter__` and `__next__` to traverse the sequence.

```python
ran = my_range(1,11)

for i in ran:
  print(i)
```

***Iterating Directly (Custom Iterable Class)***

Similar to generator function call, we can directly iterate over the `my_range` class within a `for` loop, creating a new `my_range` object for each iteration. This is less common but demonstrates direct usage.

```python
for i in my_range(1,11):
  print(i)
```

## Implementing `my_range` with a Generator
This `my_range` function now uses `yield` to create a generator version. This significantly simplifies the code compared to the class-based implementation, achieving the same iterative behavior with fewer lines.

```python
def my_range(start,end):

  for i in range(start,end):
    yield i
```

***Retrieving Values with `next()` (Generator Version)***

Demonstrates using `next()` with the generator-based `my_range` function. Notice the similarity in usage to the class-based iterator, highlighting that generators are a more concise way to achieve iteration.

```python
gen = my_range(1,11)

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

***Generator Exhaustion (`StopIteration`) (Generator Version)***

As expected, attempting to call `next()` on an exhausted generator-based `my_range` object will also raise a `StopIteration` exception.

```python
print(next(gen))
```

***Iterating with `for` Loop (Generator Object)***

This corrected cell demonstrates iterating through the generator object `gen` created from `my_range(1, 11)` using a `for` loop. This is the most common and Pythonic way to consume a generator.

```python
gen = my_range(1,11)

for i in ran:
  print(i)
```

***Iterating Directly (Generator Function Call)***

This shows that you can directly use the generator function call within a `for` loop. Each time `my_range(1,11)` is called, it creates a new generator object that the `for` loop then consumes.

```python
for i in my_range(1,11):
  print(i)
```

## Generator Expressions
Generator expressions provide a concise syntax for creating generators, similar to list comprehensions but using parentheses instead of square brackets. They are memory-efficient because they don't construct the entire sequence in memory at once.

```python
gen = (i**3 for i in range(1,11))

print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

## Benefits of using a Generator

***1. Ease of Implementation***

Generators simplify the creation of iterators. Instead of writing a full class with `__iter__` and `__next__` methods, you can often achieve the same functionality with a single generator function using `yield`.

```python
def simple_generator():
  for i in range(3):
    yield i

print("Ease of Implementation Example:")
for item in simple_generator():
  print(item)
```

***2. Memory Efficient***

Generators produce items one at a time and do not store all items in memory simultaneously. This makes them ideal for processing large datasets or infinite sequences, as they consume significantly less memory than lists or other full-in-memory collections.

```python
import sys

# List example
my_list = [i for i in range(1000000)]
print(f"Size of list (1M integers): {sys.getsizeof(my_list)} bytes")

# Generator example
my_generator = (i for i in range(1000000))
print(f"Size of generator (1M integers): {sys.getsizeof(my_generator)} bytes")
```

***3. Representing Infinite Streams***

Generators can represent potentially infinite sequences of data because they generate values on demand. This is impossible with traditional functions that return a complete list, as such a list would require infinite memory.

```python
def infinite_sequence():
  num = 0
  while True:
    yield num
    num += 1

# Using an infinite generator (be careful not to run indefinitely)
inf_gen = infinite_sequence()
print("Infinite Stream (first 5 values):")
for _ in range(5):
  print(next(inf_gen))
```

***4. Chaining Generators***

Generators can be easily chained together, where the output of one generator becomes the input of another. This allows for powerful and memory-efficient data processing pipelines, where data is streamed through several transformations without intermediate storage.

```python
def even_numbers(start, end):
  print("Generating even numbers...")
  for i in range(start, end):
    if i % 2 == 0:
      yield i

def squared_numbers(numbers_gen):
  print("Squaring numbers...")
  for num in numbers_gen:
    yield num * num

# Chaining generators: get even numbers, then square them
evens = even_numbers(1, 10)
squared_evens = squared_numbers(evens)

print("Chained Generators (Squared Evens):")
for s_even in squared_evens:
  print(s_even)
```

***5. Laziness***

Generators are inherently lazy, meaning they only compute values when requested. This can save computation time if not all values in a sequence are needed.

```python
def lazy_generator():
  print("Starting generator...")
  yield 1
  print("Generated 1, pausing...")
  yield 2
  print("Generated 2, pausing...")
  yield 3
  print("Generated 3, finishing...")

lazy_gen = lazy_generator()

print("Before first next() call.")
print(next(lazy_gen))
print("After first next() call.")

print("Before second next() call.")
print(next(lazy_gen))
print("After second next() call.")
```
