# 3. Variables

```python
# Static Vs Dynamic Typing
# Static Vs Dynamic Binding
# stylish declaration techniques
```

```python
# Static Typing (c++,java,c#...etc)
# The type of a variable is known before the program runs (at compile time).

# int a = 5;

# Dynamic Typing (python..)
# The type is decided while the program is running

name = 'sony'
print(name)

```

```python
# Dynamic Binding
a = 5
print(a)
a = 'sony'
print(a)

# Static Binding

# int a = 5
# a='sony'-->>error
```

Different type of variable creation

```python
a = 1
b = 2
c = 3
print(a,b,c)
```

```python
a,b,c = 1,2,3
print(a,b,c)
```

```python
a = b = c = 5
print(a,b,c)
```

# 4. Keywords & Identifiers

```python
# Keywords

import keyword

# List all keywords
python_keywords = keyword.kwlist
print("Python Keywords ({}):".format(len(python_keywords)))
print(python_keywords)
```

```python
# Python Identifiers Example

# Valid identifiers
name1 = 'Sony'  # Can't start with a digit
print(name1)

_ = 'Sony'       # Can use underscore(only one special character allowed)
print(_)

# Invalid identifier (can't be a keyword)
# if = 5  # ❌ This will give an error because 'if' is a keyword
```

# 7. Literals

```python
# Integer Literals
a = 0b1010  # Binary Literal
b = 100     # Decimal Literal
c = 0o310   # Octal Literal
d = 0x12c   # Hexadecimal Literal
```

```python
# Float Literals
float_1 = 10.5
float_2 = 1.5e2   # 1.5 * 10^2
float_3 = 1.5e-3  # 1.5 * 10^-3
```

```python
# Complex Literal
x = 3.14j

print(a, b, c, d)
print(float_1, float_2, float_3)
print(x, x.imag, x.real)
```

```python
# String Literals
string = 'This is Python'
strings = "This is Python"
char = "C"
multiline_str = """This is a multiline string with more than one line code."""
unicode_str = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode_str)
print(raw_str)
```

```python
# Boolean Literals
a = True + 4
b = False + 10

print("a:", a)
print("b:", b)
```

```python
# None Literal(python can't declare variable with initialization, that's why None use)
k = None
a = 5
b = 6
print('Program executed')
```
