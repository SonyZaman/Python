# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

### Creating Sets

```python
# empty
s = set()
print(type(s))

# 1D
s1 = {1,2,3,4}
print(s1)

# homogeneous
s3 = {3,6,7,1,5}
print(s3)

# heterogeneous
s4 = {1,2.5,'hello',(1,2,3)}
print(s4)

# using type conversion
s5 = set([1,5,3,6])
print(s5)

# duplicates not allowed
s6 = {1,1,2,2,3,3}
print(s6)
```

Set can't have mutable items

```python
# A set cannot contain mutable items like another set.
s1 = {1,2,3,{4,5}}

# A set cannot contain mutable items like a list.
s2 = {1,2,3,[4,5]}
```

```python
# Sets are unordered, so s1 and s2 are considered equal as they contain the same elements.
s1 = {1,2,3}
s2 = {3,2,1}

print(s1 == s2)
```

### Accessing Items

```python
s1 = {1,2,3,4}
s1[0:3]
```

### Editing Items

```python
# Individual items in a set cannot be access/edited directly with index

s1 = {1,2,3,4}
s1[0] = 100
```

### Adding Items

```python
s = {1,2,3,4}

# add
s.add(5)
print(s)
```

```python
# update
s.update((13,14))
s.update({10,11})
s.update([5,6,7])
print(s)
```

### Deleting Items

```python
# del - Deletes the entire set object. Index-wise deletion is not possible for set elements.
s = {1,2,3,4,5,7,9,10,11}
del s[0] # This would raise a TypeError: 'set' object doesn't support item deletion
```

```python
# discard() - Deletes a value. It does not raise an error if the value is not found.

s = {1,2,3,4,5,7,9,10,11}

s.discard(30)
s.discard(3)
print(s)
```

```python
# remove() - Deletes a value. It raises a KeyError if the value is not found.

s = {1,2,3,4,5,7,9,10,11}

s.remove(5)
s.remove(50) #<<--error
print(s)
```

```python
# pop() - Removes first element and returns an arbitrary element from the set.
s = {1,2,3,4,5}
print(s)

s.pop()
print(s)

s.pop()
print(s)

# clear() - Removes all elements from the set, making it an empty set.
s.clear()
print(s)
```


### Set Operation

```python
# Union (|) combines all unique elements from both sets.
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 | s2
```

```python
# Intersection (&) returns elements common to both sets.
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 & s2
```

```python
# Difference (-) returns elements in the first set but not in the second.
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

print(s1 - s2)
print(s2 - s1)
```

```python
# Symmetric Difference (^) returns elements in either set, but not in both.
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 ^ s2
```

```python
# Membership test ('not in') returns True if the element is not present.
s1 = {1,2,3,4,5}

1 not in s1
```

```python
# Iterating through the elements of a set
s1 = {1,2,3,4,5}

for i in s1:
  print(i)
```

### Set Functions

```python
s = {3,1,4,5,2,7}

# Using built-in functions: len(), sum(), min(), max(), sorted()
print(len(s))

print(sum(s))

print(min(s))

print(max(s))

print(sorted(s,reverse=True))
```

```python
# union()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# union() returns a new set with all unique elements from both sets.
print(s1.union(s2)) # Corrected to s2 for a meaningful union
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
```

```python
#  update()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# update() adds elements from another iterable to the set.
s1.update(s2) # permanent changes in s1
print(s1)
print(s2)
```

```python
# intersection()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# intersection() returns a new set with common elements.
print(s1.intersection(s2))
print(s1)
print(s2)
```

```python
# intersection_update()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# intersection_update() removes elements from the set that are not present in both.
print(s1.intersection_update(s2)) # permanent changes in s1
print(s1)
print(s2)
```

```python
# difference()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# difference() returns a new set with elements in s1 but not in s2.
s1.difference(s2)
```

```python
#difference_update()

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# difference_update() removes all elements from s1 that are also present in s2.
s1.difference_update(s2) # permanent
print(s1)
print(s2)
```

```python
# symmetric_difference()
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# symmetric_difference() returns a new set with elements in either s1 or s2 but not in both.
s1.symmetric_difference(s2)

```

```python
# symmetric_difference_update() methods
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# symmetric_difference_update() updates the set with elements in either set, but not in both.
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint()
# isdisjoint() returns True if two sets have no elements in common.
s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.isdisjoint(s2)
```

```python
# issuperset()
# issuperset() returns True if all elements of another set are present in this set.
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issuperset(s2)
```

```python
# copy()
# copy() creates a shallow copy of the set.
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
print(s2)
```

### Frozenset
Frozen set is just an immutable version of a Python set object

```python
# create frozenset
fs1 = frozenset([1,2,3])
fs2 = frozenset([3,4,5])
print(fs1)
print(fs2)
```

Set operation

```python
print(fs1 & fs2)
print(fs1 | fs2)
print(fs1 ^ fs2)
print(fs1 - fs2)
```

```python
# what works and what does not
# works -> all read functions
# does't work -> write operations
```

```python
# When to use
# 2D sets (possible, here sets contain immutable elements)
fs = frozenset([1,2,frozenset([3,4])])
fs
```
